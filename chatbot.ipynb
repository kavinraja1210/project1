```python
# ==========================================================
# RETRIEVAL-BASED CHATBOT
# ==========================================================

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ----------------------------
# Knowledge Base
# ----------------------------

questions = [
    "hello",
    "hi",
    "how are you",
    "what is your name",
    "who created you",
    "what is artificial intelligence",
    "what is machine learning",
    "what is python",
    "what is data science",
    "what is cloud computing",
    "what is internet of things",
    "what is deep learning",
    "what is computer vision",
    "what is natural language processing",
    "bye",
    "thank you"
    "hi i am akshay"
]

answers = [
    "Hello! How can I help you?",
    "Hi! Nice to meet you.",
    "I'm doing great! How about you?",
    "I am a Retrieval-Based Chatbot.",
    "I was created using Python and Machine Learning.",
    "Artificial Intelligence (AI) enables machines to perform tasks that normally require human intelligence.",
    "Machine Learning is a subset of AI where systems learn from data.",
    "Python is a popular programming language used for AI, ML, and web development.",
    "Data Science is the process of extracting useful information from data.",
    "Cloud Computing provides computing resources over the internet.",
    "The Internet of Things (IoT) connects smart devices through the internet.",
    "Deep Learning uses neural networks with multiple layers to learn patterns.",
    "Computer Vision enables computers to understand images and videos.",
    "Natural Language Processing (NLP) helps computers understand human language.",
    "Goodbye! Have a nice day.",
    "You're welcome!"
    "oh , hi akshay"
]

# ----------------------------
# TF-IDF Vectorizer
# ----------------------------

vectorizer = TfidfVectorizer()

question_vectors = vectorizer.fit_transform(questions)

print("Chatbot is Ready!")
print("Type 'exit' to stop.\n")

# ----------------------------
# Chat Loop
# ----------------------------

while True:

    user_input = input("You : ")

    if user_input.lower() == "exit":
        print("Bot : Goodbye!")
        break

    # Convert user input into vector
    user_vector = vectorizer.transform([user_input])

    # Compute similarity
    similarity = cosine_similarity(user_vector, question_vectors)

    # Best matching question
    index = similarity.argmax()

    score = similarity[0][index]

    if score < 0.20:
        print("Bot : Sorry, I don't understand that.")
    else:
        print("Bot :", answers[index])
```
