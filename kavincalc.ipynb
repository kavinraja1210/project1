```python
import tkinter as tk

root = tk.Tk()
root.title("Calculator")
root.geometry("300x400")

entry = tk.Entry(root, font=("Arial",20), justify="right")
entry.pack(fill="both", padx=10, pady=10)

def press(num):
    entry.insert(tk.END, str(num))

def equal():
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

def clear():
    entry.delete(0, tk.END)

buttons = [
    '7','8','9','/',
    '4','5','6','*',
    '1','2','3','-',
    '0','.','=','+'
]

frame = tk.Frame(root)
frame.pack()

row = 0
col = 0

for b in buttons:
    if b == "=":
        cmd = equal
    else:
        cmd = lambda x=b: press(x)

    tk.Button(frame, text=b, width=5, height=2, font=("Arial",16),
              command=cmd).grid(row=row, column=col, padx=2, pady=2)

    col += 1
    if col > 3:
        col = 0
        row += 1

tk.Button(root, text="Clear", font=("Arial",16), command=clear).pack(fill="x", padx=10, pady=10)

root.mainloop()
```
